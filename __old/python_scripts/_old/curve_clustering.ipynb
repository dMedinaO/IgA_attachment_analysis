```python
import pandas as pd

df_data = pd.read_csv("../full_analysis/data_measures.csv")
values_to_clustering = df_data.drop(columns=['id_name'])
values_to_clustering = values_to_clustering.fillna(0)
```

```python
from sklearn.cluster import MeanShift

meanshif_instance = MeanShift()
meanshif_instance.fit(values_to_clustering)

values_to_clustering['label'] = meanshif_instance.labels_
values_to_clustering
```
