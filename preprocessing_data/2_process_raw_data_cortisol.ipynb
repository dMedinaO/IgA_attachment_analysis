- Loading modules

```python
import pandas as pd
```

- Read cortisol dataset

```python
df_cortisol = pd.read_csv("../raw_data/cortisol/raw_data_measures.csv")
df_cortisol.sort_values(by="Sample")
```

```python
df_cortisol.shape
```

- Read processed IgA dataset

```python
df_processed_data = pd.read_csv("../processed_data/1_processed_data_IgA.csv")
df_processed_data.head(5)
```

- Making merge and deleting NaN values

```python
df_merge = df_cortisol.merge(right=df_processed_data, left_on="Sample", right_on="ID", how="left")
df_merge
```

```python
df_merge = df_merge.drop(columns=["age_baby_1",	"age_baby_2", "age_baby_5"])
```

```python
df_merge = df_merge.dropna()
df_merge
```

```python
df_merge.describe()
```

- Processing merge dataset

```python
df_merge.columns
```

```python
df_merge = df_merge[['ID', 'Cortisol (pg/mL)', 'Gender',
       'sIgA_Average_Morning', 'sIgA_Average_Afternoon',
       'sIgA_Average_Difference', 'sIgA_Average_Morning_log',
       'sIgA_Average_Afternoon_log', 'sIgA_Average_Difference_log',
       'Attachment_y']]
df_merge.columns = ['ID', 'Cortisol (pg/mL)', 'Gender',
       'sIgA_Average_Morning', 'sIgA_Average_Afternoon',
       'sIgA_Average_Difference', 'sIgA_Average_Morning_log',
       'sIgA_Average_Afternoon_log', 'sIgA_Average_Difference_log',
       'Attachment']
df_merge
```

- Checking some statistics

```python
df_merge[["Cortisol (pg/mL)", "Attachment"]].groupby(by=["Attachment"]).mean()
```

```python
df_merge[["Cortisol (pg/mL)", "Attachment"]].groupby(by=["Attachment"]).median()
```
