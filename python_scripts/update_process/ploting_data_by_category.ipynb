```python
for colum_apego in columns_apego:
    print("Processing column: ", colum_apego)

    command = f"mkdir -p ../../results_update_process/boxplots/{colum_apego}"
    os.system(command=command)

    for column in ['ADS_CHILD_M', 'ADS_CHILD_V', 'ADS_CHILD_TA', 'ADS_CHILD_SA',
        'ADS_CHILD_A', 'ADS_CHILD_P', 'ADS_MAIN_CAREGIVER_M',
        'ADS_MAIN_CAREGIVER_V', 'ADS_MAIN_CAREGIVER_TA',
        'ADS_MAIN_CAREGIVER_SA', 'ADS_MAIN_CAREGIVER_A', 'ADS_MAIN_CAREGIVER_P','change_attachment_y', 
        'sensitivity2', 'Rango_ITERS_y', 'insecure_bebe2', 'Género_bebé_y',
        'Grupo_étnico_bebé_y']:
    
        fig, axes = plt.subplots(3, 3, figsize=(40, 60), sharex=True)

        sns.boxplot(ax=axes[0][0], data=df_data, x=column, hue=colum_apego, y="m1", fill=False)
        sns.boxplot(ax=axes[0][1], data=df_data, x=column, hue=colum_apego, y="m2", fill=False)
        sns.boxplot(ax=axes[0][2], data=df_data, x=column, hue=colum_apego, y="m3", fill=False)
        sns.boxplot(ax=axes[1][0], data=df_data, x=column, hue=colum_apego, y="m4", fill=False)
        sns.boxplot(ax=axes[1][1], data=df_data, x=column, hue=colum_apego, y="average_d1", fill=False)
        sns.boxplot(ax=axes[1][2], data=df_data, x=column, hue=colum_apego, y="average_d2", fill=False)
        sns.boxplot(ax=axes[2][0], data=df_data, x=column, hue=colum_apego, y="average_m", fill=False)
        sns.boxplot(ax=axes[2][1], data=df_data, x=column, hue=colum_apego, y="average_t", fill=False)

        fig.autofmt_xdate(rotation=45)
        
        plt.tight_layout()
        
        name_fig = f"../../results_update_process/boxplots/{colum_apego}/{column}.png"
        plt.savefig(name_fig)

        plt.clf()
        
```
