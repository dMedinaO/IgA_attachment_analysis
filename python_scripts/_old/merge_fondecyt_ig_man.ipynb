```python
import pandas as pd

data_fondecyt = pd.read_excel("../data_to_work_dm_12042022/180110 data FONDECYT_iniciacion (1).xlsx")
data_ig_man = pd.read_excel("../data_to_work_dm_12042022/Base de datos 13.10.2017 - _MAN_actualizado 19.10.17.xlsx", sheet_name="Update-2", skiprows=1)
```

```python
#id column with baby name for ig dataset
array_name = []

for index in data_ig_man.index:
    id_name = "{}-{}-{}".format(data_ig_man['NOMBRE'][index], data_ig_man['APELLIDO PATERNO'][index], data_ig_man['APELLIDO MATERNO'][index])
    array_name.append(id_name)

data_ig_man['id_name'] = array_name
```

```python
#id column with baby name for fondecyt dataset

array_name = []

for index in data_fondecyt.index:
    id_name = "{}-{}-{}".format(data_fondecyt['Nombre_bebé'][index], data_fondecyt['Apellido1_bebé'][index], data_fondecyt['Apellido2_bebé'][index])
    array_name.append(id_name)

data_fondecyt['id_name'] = array_name

```

```python
#create dataframe with relevant measures -> in this case, it is relevant the column used to evaluate the babies
data_measures = []

column_to_check = "[ug/ml]"

data_none = [None for i in range(6)]

for unique_baby in data_ig_man['id_name'].unique():
    df_filter = data_ig_man.loc[data_ig_man['id_name'] == unique_baby]
    df_filter = df_filter.reset_index()

    row = [unique_baby] + data_none
    
    for index in df_filter.index:
        measure = df_filter[column_to_check][index]
        pos_measure = int(df_filter['ID MUESTRA'][index].split("-")[1])
        row[pos_measure] = measure
    data_measures.append(row)

data_measures_df = pd.DataFrame(data_measures, columns=['id_name', 'm1', 'm2', 'm3', 'm4', 'm5', 'm6'])
data_measures_df
```

```python
data_measures_df.to_csv("../full_analysis/data_measures.csv", index=False)
```

```python
#remove baybies
filter_data = []

for index in data_measures_df.index:
    row = data_measures_df.iloc[index].to_list()
    id_name = row[0]
    measures = row[1:]

    count_measures =0
    for measure in measures:
        if str(measure) != 'nan' and measure != 0:
            count_measures +=1
    
    if count_measures == 4:
        measures_value = [value for value in measures if str(value) != 'nan' and value != 0]
        measures_value.insert(0, id_name)
        filter_data.append(measures_value)

df_filter_measures = pd.DataFrame(filter_data, columns=['id_name', 'm1', 'm2', 'm3', 'm4'])
df_filter_measures
```

```python
df_filter_measures = pd.read_csv("../full_analysis/data_measures.csv")
df_filter_measures
```

```python
#simple plot of measures by point
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-white")
plt.rc('axes', grid=False, facecolor="white")
plt.rcParams.update({'font.size': 14})

fig, ax = plt.subplots(figsize=(17, 10), layout='constrained')

categories = ['m1', 'm2', 'm3', 'm4']

for index in df_filter_measures.index:
    data_values = df_filter_measures.iloc[index].to_list()
    ax.plot(categories, data_values[1:],  label=data_values[0])  # Plot some data on the axes.

plt.xlabel('Measures')
plt.ylabel(column_to_check)
plt.legend() 
```

```python
#make histogram by measure
df_list = []

for measure in ['m1', 'm2', 'm3', 'm4']:
    df_values = df_filter_measures[['id_name', measure]]
    df_values = df_values.rename(columns={measure: column_to_check})
    df_values['measure'] = measure
    df_list.append(df_values)

df_to_boxplot = pd.concat(df_list, axis=0)
df_to_boxplot = df_to_boxplot.reset_index()

sns.boxplot(data=df_to_boxplot, x=column_to_check, y='measure')
```

```python
#merge data with fondecyt
data_merge = pd.merge(df_filter_measures, data_fondecyt, on="id_name")
data_merge
```

```python
#make histogram by measure
df_list = []

for measure in ['m1', 'm2', 'm3', 'm4']:
    df_values = data_merge[['id_name', measure, 'attachment2_ads', 'Attachment_ads', 'change_attachment', 'poschange_attachment', 'negchange_attachment', 'Edad_bebé_1']]
    df_values = df_values.rename(columns={measure: column_to_check})
    df_values['measure'] = measure
    df_list.append(df_values)

df_to_boxplot = pd.concat(df_list, axis=0)
df_to_boxplot = df_to_boxplot.reset_index()


fig, axes = plt.subplots(2, 3, sharex=True, sharey=True, figsize=(20, 10))

i = 0
j = 0
index =0

for column in ['attachment2_ads', 'Attachment_ads', 'change_attachment', 'poschange_attachment', 'negchange_attachment', 'Edad_bebé_1']:
    index+=1
    sns.boxplot(ax=axes[i][j], data=df_to_boxplot, x=column_to_check, y='measure', hue=column)

    if index == 3 or index==6:
        i += 1
        j = 0
    else:
        j+=1

plt.savefig("demo_comparison.pdf")

```
